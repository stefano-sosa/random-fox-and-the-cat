# Random Fox API

```python
import requests
```

```python
import json
```

```python
from io import BytesIO
```

```python
from PIL import Image
```

```python
url = 'https://randomfox.ca/floof/'
```

```python
req = requests.get(url)
```

```python
load = json.loads(req.content)
```

```python
imageurl = load['image']
```

```python
imagereq = requests.get(imageurl)
```

```python
Image.open(BytesIO(imagereq.content))
```
